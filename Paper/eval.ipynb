```python
import os
import csv
import cv2
import numpy as np
import pandas as pd
import json 

# Plotting
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
def plotLossHistory(loss_history, validation_loss, name):
    """
    Plots training history

    """
    fontsize = 12

    %matplotlib inline
    plt.figure(dpi=90)
    plt.semilogy(loss_history)
    plt.semilogy(validation_loss)
    plt.title("Training history of {}".format(name), fontsize=fontsize)
    plt.xlabel("Epoch", fontsize=fontsize)
    plt.ylabel("Loss", fontsize=fontsize)
    plt.legend(['loss', 'validation_loss'], fontsize=fontsize)

    # Save figure
#     filename = "figures/traning_hist_" + name
#     plt.savefig(filename)
    plt.show()
```

```python
df_cnn = pd.read_csv("models/model_4_loss_history.csv")
print(df_cnn.shape)
df_cnn.head()
```

```python
plotLossHistory(df_cnn["loss"],df_cnn["cal_loss"], "CNN")
```

```python
df_mlp = pd.read_json("models/report_params.json")
print(df_mlp.shape)
df_mlp.head()
```

```python
plotLossHistory(df_mlp["loss_history"][2],df_mlp["validation_loss"][2], "MLP")
```
